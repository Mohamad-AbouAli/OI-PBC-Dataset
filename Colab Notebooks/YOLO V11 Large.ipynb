```python
from ultralytics import YOLO

# Load YOLO model (ensure correct name)
model = YOLO("yolo11l.pt")  # Replace with your model
```

✅ 5️⃣ Train YOLOv8

```python
# Train and enable logging
results = model.train(
    data="/content/drive/MyDrive/PBC_YOLO - 721/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    lr0=0.01,
    lrf=0.0001,
    momentum=0.937,
    weight_decay=0.0005,
    project='/content/drive/MyDrive/SavedModel/YOLO',  # Save directory
    name='YOLO11Large721-Auto-Epoch100v4 Train-Val',  # Name for the run
    save=True,
    workers=8,
    patience=15,
    cos_lr = True,
    visualize = True,
    save_dir = "/content/drive/MyDrive/SavedModel/YOLO/YOLO11Large721",
    plots=True  # Automatically saves training curves
)
```

**Saving Model**
